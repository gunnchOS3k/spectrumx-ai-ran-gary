```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
DATA_DIR = "competition_dataset/files/..."

labels_df = pd.read_csv("groundtruth.csv")

scores_energy = []
scores_norm = []
scores_kurt = []
labels = []

for _, row in labels_df.iterrows():
    file_path = os.path.join(DATA_DIR, row["filename"])
    iq = np.load(file_path)

    scores_energy.append(energy_detector(iq))
    scores_norm.append(normalized_energy_detector(iq))
    scores_kurt.append(kurtosis_detector(iq))
    labels.append(row["label"])
```

```python
fpr, tpr, thresholds = roc_curve(labels, scores_energy)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"Energy Detector ROC (AUC = {roc_auc:.3f})")
plt.show()
```
